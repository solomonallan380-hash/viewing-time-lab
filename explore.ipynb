## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
print("Control length:", len(control_clean))
print("Treatment length:", len(treatment_clean))

print("Control data:", control_clean[:10])
print("Treatment data:", treatment_clean[:10])
```

```python
print(control_raw)
print(treatment_raw)

print("Control raw length:", len(control_raw))
print("Treatment raw length:", len(treatment_raw))
```

```python
# TODO: Calculate the standard deviation of both the control and treatment group
import statistics

control_std = statistics.stdev(control_clean)
treatment_std = statistics.stdev(treatment_clean)

# TODO: print your findings
print("Control Standard Deviation:", control_std)
print("Treatment Standard Deviation:", treatment_std)
```

```python
# TODO: Calculate the mean of both the control and treatment group
control_mean = statistics.mean(control_clean)
treatment_mean = statistics.mean(treatment_clean)


# TODO: print your findings
print("Control Mean:", control_mean)
print("Treatment Mean:", treatment_mean)
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
outliers_control = []
lower_bound = control_mean - 2 * control_std
upper_bound = control_mean + 2 * control_std

for value in control_clean:
    if value < lower_bound or value > upper_bound:
        outliers_control.append(value)

print(outliers_control)
```

```python
# TODO: Are there any outliers in your treatment group?
outliers_treatment = []
lower_bound = treatment_mean - 2 * treatment_std
upper_bound = treatment_mean + 2 * treatment_std

for value in treatment_clean:
    if value < lower_bound or value > upper_bound:
        outliers_treatment.append(value)

print(outliers_treatment)
```
